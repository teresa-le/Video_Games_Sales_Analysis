```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

vgdata = pd.read_csv('../DATA/clean_videogamesalesREAL.csv')
vgdata.head()
```

```python
vgdata2 = vgdata.copy()
```

```python
vgdata2.head()
```

```python
vgdata2 = vgdata2.drop(['Name'], axis = 1)
```

```python
vgdata2.head()
```

```python
vgdata2.head()
```

```python
rat_counts = vgdata2['Rating'].value_counts()
rat_counts
```

```python
# Determine which values to replace if counts are less than ...?
replace_rating = list(rat_counts[rat_counts < 1000].index)

# Replace in dataframe
for rat in replace_rating:
    vgdata2.Rating = vgdata2.Rating.replace(rat,"Other")
    
# Check to make sure binning was successful
vgdata2.Rating.value_counts()
```

```python
vgdata2.head()
```

```python
vgdata2 = vgdata2[['Platform','Publisher','NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales','Years_on_Market','Critic_Score', 'User_Score','Genre', 'Rating']]
```

```python
vgdata2.head()
```

```python
vgdata2 = vgdata2.drop(['Platform','Publisher'], axis = 1)
```

```python
# Generate our categorical variable lists
vgsales_cat = vgdata2.dtypes[vgdata2.dtypes == "object"].index.tolist()
print(vgsales_cat)
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(vgdata2[vgsales_cat]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(vgsales_cat)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
vgdataCL = vgdata2.merge(encode_df,left_index=True, right_index=True)
vgdataCL = vgdataCL.drop(vgsales_cat,1)
vgdataCL.head()
```

```python
vgdataCL.to_csv('../DATA/regreadyvgsales3.csv', index = False, encoding = "UTF8")
```

```python
vgdataCL['Publisher'] = vgdata['Publisher']
vgdataCL['Platform'] = vgdata['Platform']
```

```python
vgdataCL.head()
```

```python
plat_counts = vgdataCL['Platform'].value_counts()
plat_counts
```

```python
# Determine which values to replace if counts are less than ...?
replace_platform = list(plat_counts[plat_counts < 150].index)

# Replace in dataframe
for plat in replace_platform:
    vgdataCL.Platform = vgdataCL.Platform.replace(plat,"Other")
    
# Check to make sure binning was successful
vgdataCL.Platform.value_counts()
```

```python
pub_counts = vgdataCL['Publisher'].value_counts()
pub_counts
```

```python
# Determine which values to replace if counts are less than ...?
replace_publisher = list(pub_counts[pub_counts < 10].index)

# Replace in dataframe
for pub in replace_publisher:
    vgdataCL.Publisher = vgdataCL.Publisher.replace(pub,"Other")
    
# Check to make sure binning was successful
vgdataCL.Publisher.value_counts()
```

```python
# Generate our categorical variable lists
vgsales2_cat = vgdataCL.dtypes[vgdataCL.dtypes == "object"].index.tolist()
print(vgsales2_cat)
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df2 = pd.DataFrame(enc.fit_transform(vgdataCL[vgsales2_cat]))

# Add the encoded variable names to the dataframe
encode_df2.columns = enc.get_feature_names(vgsales2_cat)
encode_df2.head()
```

```python
vgdataCL2 = vgdataCL.merge(encode_df2,left_index=True, right_index=True)
vgdataCL2 = vgdataCL2.drop(vgsales2_cat,1)
vgdataCL2.head()
```
